Aprendizaje Supervisado

```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
#Carga del conjunto de datos sobre el Titanic
titanic = sns.load_dataset('titanic')
```

```python
titanic.head(10)
```

```python
titanic.info()
```

```python
titanic.tail()
```

```python
#Eliminar datos nulos
titanic.dropna(inplace=True)
```

```python
#Seleccionar características y objetivos
x = titanic[['pclass','sex','age']] #características
y = titanic['survived'] #objetivo
```

```python
#Convertir la variable 'sex' (categórica) en una variable numérica
x['sex'] = x['sex'].map({'male':0,'female':1}) 
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
prediction = model.predict(x_test)
```

```python
print('Accuracy: ', accuracy_score(y_test,prediction))
```

```python
print('Confusion Matrix: ',confusion_matrix(y_test,prediction))
```

```python
print('classification Report: ',classification_report(y_test, prediction))
```

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from sklearn import datasets
```

```python
#Carga del conjunto de datos
wine = datasets.load_wine()
x = wine.data
```

```python
# Normalización de los datos ya que K-Means es sensible a la escala
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
# Entrenamiento del model K-Means
kmeans = KMeans(n_clusters=3,random_state=42)
kmeans.fit(x)
```

```python
#Predicción de Clusters
labels = kmeans.predict(x)
```

```python
#Calcular puntuación de silueta para evaluar el rendimiento
silhouette_avg = silhouette_score(x,labels)
print('El puntaje de silueta es: ', silhouette_avg)
```

```python
#Para una comparación más avanzada, se puede comparar con las etiquetas reales
accuracy = np.mean(labels == wine.target)
print('La precisión comparando con las etiquetas reales es: ', accuracy)
```
